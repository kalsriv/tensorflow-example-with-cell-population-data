```python
from sklearn.compose import make_column_transformer
```

```python
column_trans =  make_column_transformer(
    (OneHotEncoder(), ['Status']),
    remainder = 'passthrough')
```

```python
column_trans.fit_transform(df)
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(column_trans, logreg)
```

# Trying it on linear regression

```python
df = pd.read_csv('/home/srivastava/Documents/sklearn_work/NK_Cells.csv')
```

```python
df = df.drop('Donor', axis='columns')
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()


# 2/3. FIT AND TRANSFORM
# use df.apply() to apply le.fit_transform to all columns
X_2 = df.apply(le.fit_transform)
X_2.head()

```

```python
X = X_2.drop('Status', axis='columns')
y=X_2.Status
```

```python
X_2_new = X_2.sample(5, random_state=99)
X_new = X_2_new.drop('Status', axis='columns')
y_new = X_2_new.Status
```

```python

from sklearn.linear_model import LinearRegression
from sklearn import linear_model
# Create linear regression object
regr = linear_model.LinearRegression()

X_train = X[:-20]
X_test = X[-20:]

# Split the targets into training/testing sets
y_train = y[:-20]
y_test = y[-20:]

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)
y_pred
```

```python
y_test
```

# KNN methods for prediction

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=20)

# Train the model using the training sets
model.fit(X_train, y_train)

#Predict Output
predicted= model.predict(X_new) 
print(predicted)
```
